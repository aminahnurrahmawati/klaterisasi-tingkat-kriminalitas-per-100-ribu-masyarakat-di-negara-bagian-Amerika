```python
#cek outlier

def finding_outlier(data): #batas bawah
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3-Q1
    df_final=data[(data<(Q1-(1.5*IQR)))]
    return df_final
```

```python
finding_outlier(data) #mencari outlier dari setiap data
```

```python
def finding_outlier2(data):#batas atas
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3-Q1
    df_final2=data[(data>(Q3+(1.5*IQR)))]
    return df_final2
```

```python
finding_outlier2(data)
```

Didapatkan outlier yang bernilai di atas kuartil ke tiga pada data Robbery pada negara bagian NY dan VT dan data murder pada negara bagian VT. hal ini merepresentasikan bahwa pada kasus-kasus tersebut melunjak tinggi pada negara-negara bagian NY dan VT jika dibandingkan dengan negara bagian lain

# PCA and Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, normalize
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
data_to_process = data.drop('Unnamed: 0', axis=1) #karna akan melakukan clustering, maka kita dapat me-remove primary key (nama negara)
```

# Standardize and Normalize

```python
scaler = StandardScaler() 
scaled_data = scaler.fit_transform(data_to_process) 
norm_data = normalize(scaled_data)
```

# PCA

```python
pca = PCA()
pca.fit(norm_data)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
#PCA Summary

n_components=2
pca_final = PCA(n_components=n_components)
pca_final.fit(norm_data)
pca_data = pca_final.fit_transform(norm_data)
```

```python
plt.scatter(pca_data[:,0], pca_data[:,1])
```

```python
for i in np.arange(n_components):
    index =  np.argmax(np.absolute(pca_final.get_covariance()[i]))
    max_cov = pca_final.get_covariance()[i][index]
    column = data_to_process.columns[index]
    print("Principal Component", i+1, "maximum covariance :", "{:.2f}".format(max_cov), "from column", column)
```

# Interpretasi PCA

Karena komponen utama adalah kombinasi linier dari atribut, pada dasarnya kita mencoba untuk mengidentifikasi variabel mana yang paling mempengaruhi komponen utama. Lebih khusus lagi, kita melihat matriks kovarians untuk melihat atribut mana yang memiliki kovarians paling banyak dengan komponen utama.

Untuk komponen utama pertama, kolom "Murder" memiliki kovarians paling banyak. Kita bisa menganggap komponen utama pertama sebagai perwakilan dari "Murder".

Untuk komponen utama kedua, kolom "Rape" memiliki kovarians paling banyak. komponen utama kedua adalah perwakilan dari "Rape".

# K-Means dari data yang telah diproses menggunakan PCA

Untuk menggunakan algoritme Kmeans, kita harus menyediakan jumlah cluster yang kita coba untuk segmentasikan data. Karena kita tidak tahu apa yang ideal, kita akan coba rentang 1-10 cluster. Kita akan melihat plot metrik untuk menentukan mana yang terbaik.

```python
sse = {}
n_clust = np.arange(2,11)

for i in n_clust:
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(pca_data)
    sse[i] = kmeans.inertia_
```

```python
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of Clusters")
plt.ylabel("Within Cluster Sum-of-Squares")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(pca_data)
```

```python
plt.scatter(pca_data[:,0], pca_data[:,1],
            c = KMeans(n_clusters = 3).fit_predict(pca_data),
            cmap = plt.cm.summer)
plt.xlabel("PC1")
plt.ylabel("PC2")

plt.show() 
```

```python
h = .01

x_min, x_max = pca_data[:,0].min() - 1, pca_data[:,0].max() + 1
y_min, y_max = pca_data[:,1].min() - 1, pca_data[:,1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = kmeans.predict(np.array(list(zip(xx.ravel(), yy.ravel()))))

Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.summer,
           aspect='auto', origin='lower')

plt.plot(pca_data[:,0], pca_data[:,1], 'k.', markersize=2)

centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='o', s=10, linewidths=3,
            color='w', zorder=10)

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

plt.xlabel("PC1")
plt.ylabel("PC2")

plt.show()
```

```python
for i in np.arange(len(centroids)):
    print("Center of Cluster", i+1, ":", centroids[i])
```

# K-Means Clustering

untuk klasterisasi kali ini, kita menggunakan K-means Clustering tanpa PCA terlebih dahulu. 
kita mengklasterisasi setiap variable kriminal 
sehingga untuk satu kolom (jenis kriminal) kita akan membagi menjadi beberapa klaster

Pada pengaplikasian K-means Clustering, kita mencari terlebih dahulu banyak kluster (k) berdasarkan data yang ada dengan menggunakan "Elbow Method". Nilai k diambil dari 1 sampai 10, nilai k diambil pada satu titik yang menujam tajam setelah diplot.

Kita mencari nilai WCSS (Within-Cluster Sum of Square) yaitu kuadrat jumlah dari jarak masing-masing poin/objek terhadap sentroid

Untuk nilai WCSS setiap tipe kriminalitas yang diklasterisasi akan ditampilkan pada codes di bawah

1. Cluster untuk Kasus "Murder"

```python
Murder = data.iloc[:,[1]]
```

```python
Murder.head()
```

```python
Murder.shape
```

```python
#Clustering with KMeans

from sklearn.cluster import KMeans
wcss=[]

for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(Murder)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('n_cluster')
plt.ylabel('wcss')

plt.show()
```

```python
wcss
```

```python
kmeans1 = KMeans(n_clusters = 3, init='k-means++', random_state=42).fit(Murder)
labels1 = kmeans1.labels_
labels1
```

```python
#mengklasifikasi setiap data berada pada klaster berapa
Murder = pd.DataFrame({"State" : data['Unnamed: 0'], "Murder" : Murder['Murder'], "klaster": labels1})
Murder
```

```python
#Pusat klaster

kmeans1.cluster_centers_ 
```

# Hierarchical Clustering

untuk percobaan terakhir, kita menggunakan Hierarchial clustering. Dimana, konsep dasarnya adalah menemukan kemiripan (simmilarity) antar elemen cluster dengan prinsip nearest neighbor dengan menggunakan jarak euclidean.

pertama kita bentuk suatu hirarki (tree) yang disebut dendogram, dari sini kita bisa menentukan berapa banyak klaster yang akan digunakan

```python
X = data.iloc[:, [1, 7]].values #mengambil keseluruhan data
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Tipe Kriminal')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
#dari dendogram, kita ambik  klaster sebanyak 5 untuk menentukan tingkat kriminalitas berdasarkan data kriminal yang ada
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```python
y_hc
```

```python
prediction_klaster =pd.DataFrame({"State" : data['Unnamed: 0'], "klaster kriminalitas": y_hc})
```

```python
#data pengelompokkan hasil  klaster dimasukkan ke dalam tabel berikut
prediction_klaster
```

Summary :

nilai 0 hingga 4 merepresentasikan tingkat kriminalitas (0=sangat rendah, 1=rendah, 2=sedang, 3=tinggi, 4=sangat tinngi) di setiap negara bagian berdasarkan data tingkat kriminal yang terjadi pada negara tersebut.
